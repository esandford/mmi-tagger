```python
from __future__ import (absolute_import, division, print_function,
                        unicode_literals)
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import corner
import copy
import pickle
from astropy.io import ascii

%matplotlib inline
```

```python
def softmax(x,b):
    return np.exp(b*x)/np.sum(np.exp(b*x))
```

```python
x = np.random.randint(0,100,50)
plt.plot(x,'bo',alpha=0.5)
plt.show()

print(softmax(x,b=2.))
```

```python
print(np.sum(softmax(x,2)))

fig = plt.plot(softmax(x,2),'bo',alpha=0.5)
fig = plt.plot(softmax(x,0.5),'ro',alpha=0.5)
#plt.yscale("log")
plt.show()
```
